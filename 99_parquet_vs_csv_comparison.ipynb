# Benchmark: Parquet vs CSV Performance


```python
import pandas
import time
import pathlib

PARQUET_PATH = list(pathlib.Path("data/processed").glob("*.parquet"))[0]
TEMP_CSV_PATH = pathlib.Path("data/processed/benchmark_temp.csv")

print(f"Target Parquet file: {PARQUET_PATH}")
```

## 1. Timing CSV Performance
We first convert the data to CSV and measure the time it takes to read it back into memory.

```python
df = pandas.read_parquet(PARQUET_PATH)
df.to_csv(TEMP_CSV_PATH, index=False)

start = time.time()
df_csv = pandas.read_csv(TEMP_CSV_PATH, low_memory=False)
csv_time = time.time() - start

print(f"CSV Read Time: {csv_time:.2f} seconds")
```

## 2. Timing Parquet Performance
Now we measure the time to read the same data from the native Parquet format.

```python
start = time.time()
df_parquet = pandas.read_parquet(PARQUET_PATH)
parquet_time = time.time() - start

print(f"Parquet Read Time: {parquet_time:.2f} seconds")
```

## 3. Results & Conclusion

```python
speedup = csv_time / parquet_time
size_csv = TEMP_CSV_PATH.stat().st_size / (1024 * 1024)
size_parquet = PARQUET_PATH.stat().st_size / (1024 * 1024)

print(f"Speedup: {speedup:.1f}x")
print(f"CSV Size: {size_csv:.2f} MB")
print(f"Parquet Size: {size_parquet:.2f} MB")
print(f"Storage Savings: {((size_csv - size_parquet) / size_csv) * 100:.1f}%")

# Cleanup
TEMP_CSV_PATH.unlink()
```
