# 03c. EDA - Call Profiles


```python
import pandas as pd
import matplotlib.pyplot as plt
import pathlib

PROCESSED_DATA_PATH = pathlib.Path("data/processed")
data_file = list(PROCESSED_DATA_PATH.glob("*_processed.parquet"))[0]
df = pd.read_parquet(data_file)

print(f"Loaded {len(df):,} records.")
```

## 1. Call Type Analysis
Analyze the most frequent reasons for police dispatch.

```python
top_types = df["CALL_TYPE"].value_counts().head(20)

plt.figure(figsize=(10, 8))
top_types.sort_values().plot(kind="barh", color="skyblue")
plt.title("Top 20 Call Types")
plt.xlabel("Frequency")
plt.show()
```

## 2. Final Dispositions
Visualize the breakdown of how calls are resolved.

```python
dispo_counts = df["FINAL_DISPO"].value_counts().head(15)

plt.figure(figsize=(10, 6))
dispo_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Top 15 Final Dispositions")
plt.ylabel("")
plt.show()

display(dispo_counts)
```
